```python
import pandas as pd
```

```python
db_covid = pd.read_csv("C:/Users/vince/OneDrive/Desktop/230520_Consegna phyton/owid-covid-data.csv", encoding="latin1", sep=",", low_memory=False)
```

```python
# 1.1verifico le dimensioni del dataset

print(len(db_covid))

print(db_covid.shape)
```

```python
# 1.2 Verifico le diciture presenti nell'intestazione

db_covid.head()
```

```python
#verifico l'elenco dei continenti tramite set

elenco_continenti = set(db_covid.continent)
print(elenco_continenti)
```

```python
#definisco un dataset semplificato, riportando le sole colonne interessate dai prossimi passaggi
db_covid_clean = db_covid.loc[ :, ["continent", "location", "total_cases", "total_vaccinations"] ]
db_covid_clean.head()
```

```python
#definisco un elenco di nazioni, da utilizzare per i prossimi passaggi

elenco_nazioni = set(db_covid_clean.location)
```

```python
#verifico il numero complessivo di casi per nazione

for nazione in elenco_nazioni:
    condizione = db_covid_clean.loc[:, "location"] == nazione
    casi_totali_nazione = db_covid_clean.loc[condizione, "total_cases"]
    print("casi totali in", nazione, "=", casi_totali_nazione.max())
```

```python
# 2. calcolo il numero totale di casi per continente

casi_totali = 0
for continente in elenco_continenti:
    casi_totali_continente = 0
    condizione = db_covid_clean.loc[:, "continent"] == continente
    db_covid_continente_clean = db_covid_clean.loc[condizione]
    for nazione in elenco_nazioni:
        if nazione != continente:
            condizione = db_covid_continente_clean.loc[:, "location"] == nazione
            casi_totali_nazione = db_covid_continente_clean.loc[condizione, "total_cases"]
            if casi_totali_nazione.max() > 0:
                casi_totali_continente += casi_totali_nazione.max()
    print("il numero totale di casi in", continente, "Ã¨ uguale a", casi_totali_continente)
    casi_totali += casi_totali_continente
print("il numero di casi totali Ã¨ ", casi_totali)
```

```python
# 3. definisco una funzione come richiesto dal punto 3 della consegna

def confronto_casi(dataset, continente1, continente2):
    continenti_draft = [continente1, continente2] 
    condizione_1 = dataset.continent == continenti_draft[0]
    condizione_2 = dataset.continent == continenti_draft[1]
    prova_1 = dataset.loc[condizione_1, "total_cases"]
    prova_2 = dataset.loc[condizione_2, "total_cases"]
    minimo_c = {continenti_draft[0]: prova_1.min(), continenti_draft[1]: prova_2.min() }
    massimo_c = {continenti_draft[0]: prova_1.max(), continenti_draft[1]: prova_2.max()}
    media_c ={continenti_draft[0]: prova_1.mean(), continenti_draft[1]: prova_2.mean()}
    percentuale_c = {continenti_draft[0]: 100*prova_1.max()/casi_totali, continenti_draft[1]: 100*prova_2.max()/casi_totali}
    s_1 = pd.Series(minimo_c)
    s_2 = pd.Series(massimo_c)
    s_3 = pd.Series(media_c)
    s_4 = pd.Series(percentuale_c)
    tabella_draft = pd.DataFrame({"minimo": s_1, "massimo":s_2, "media":s_3, "percentuale [%]":s_4})
    print(tabella_draft)
```

```python
# eseguo un test sulla funzione definita

confronto_casi(db_covid, "Africa","Asia")
```

```python
# 4. definisco una funzione come richiesto dal punto 4 della consegna

vaccinazioni_totali = 0
for continente in elenco_continenti:
    vaccini_totali_continente = 0
    condizione = db_covid.loc[:, "continent"] == continente
    db_covid_continente = db_covid.loc[condizione]
    for nazione in elenco_nazioni:
        if nazione != continente:
            condizione = db_covid_continente.loc[:, "location"] == nazione
            vaccini_totali_nazione = db_covid_continente.loc[condizione, "total_vaccinations"]
            if vaccini_totali_nazione.max() > 0:
                vaccini_totali_continente += vaccini_totali_nazione.max()
    vaccinazioni_totali += vaccini_totali_continente
print(vaccinazioni_totali)

def confronto_vaccinazioni(dataset, continente1, continente2):
    continenti_draft = [continente1, continente2] 
    condizione_1 = dataset.continent == continenti_draft[0]
    condizione_2 = dataset.continent == continenti_draft[1]
    prova_1 = dataset.loc[condizione_1, "total_vaccinations"]
    prova_2 = dataset.loc[condizione_2, "total_vaccinations"]
    minimo_c = {continenti_draft[0]: prova_1.min(), continenti_draft[1]: prova_2.min() }
    massimo_c = {continenti_draft[0]: prova_1.max(), continenti_draft[1]: prova_2.max()}
    media_c ={continenti_draft[0]: prova_1.mean(), continenti_draft[1]: prova_2.mean()}
    percentuale_c = {continenti_draft[0]: 100*prova_1.max()/vaccinazioni_totali, continenti_draft[1]: 100*prova_2.max()/vaccinazioni_totali}
    s_1 = pd.Series(minimo_c)
    s_2 = pd.Series(massimo_c)
    s_3 = pd.Series(media_c)
    s_4 = pd.Series(percentuale_c)
    tabella_draft = pd.DataFrame({"minimo": s_1, "massimo":s_2, "media":s_3, "percentuale [%]":s_4})
    print(tabella_draft)
```

```python
# eseguo un test sulla funzione definita

confronto_vaccinazioni(db_covid, "North America","South America")
```

```python
print("casi")
confronto_casi(db_covid, "Europe","South America")
confronto_casi(db_covid, "Oceania", "North America")
for i in range(0,3):
    print(" ")
   
print("vaccinazioni")
confronto_vaccinazioni(db_covid, "Europe","South America")
confronto_vaccinazioni(db_covid, "Oceania", "North America")
```

### 5. report riassuntivo
Le funzioni definite evidenziano i valori minimo, massimo, la media e la percentuale di, rispettivamentem casi totali o vaccinazioni, per i continenti richiesti. 
Il valore di minimo e di massimo rappresentano il numero minimo e massimo di casi registrati, o vaccini, tra le nazioni dell'area geografica selezionata. 
Il valore di media rappresenta la media dei casi registrati per le singole nazioni dell'area geografica selezionata.
Il valore di percentuale rappresenta il valore di casi, o di vaccinazioni, totali registrate per la nazione con il maggior numero di casi, o vaccinazioni, fratto il numero totale di casi, o vaccinazioni, registrate a livello mondiale.

Dal confronto tra i dati di Europa, America e Oceania emerge che:
* in ciascuna nazione appartenente alle nazioni delle aree geografiche indagate risulta essere presente almeno un caso di covid
* il numero massimo di casi tra le nazioni indagare risulta essere in una nazione del nord america. Al secondo posto, per nazione singola con maggior numero di casi, risulta essere l'europa, a seguire sud america e oceania
* allo stesso modo, il valore medio massimo di contagi risulta essere in nord america. Diversamente dal numero massimo, invece, l'area geografica posizionata al secondo posto per numero medio di contagi nazionali risulta essere l'oceania, al terzo posto il sud america e al quarto posto l'europa.
* Contrariamente al numero di casi, risultano esserci delle nazioni per le quali risultano essere registrati zero vaccini. 
* il continente con la nazione con il numero di vaccini maggiore risulta essere il nord america. al secondo posto troviamo il sud america, al terzo l'europa e in fondo l'oceania. 
* La media di dati di vaccinazioni risulta essere piÃ¹ alta in nord america, al secondo posto risulta il sud ameria, al terzo posto l'europa e al quarto posto l'oceania.

L'oceania risulta quindi essere l'area geografica meno coinvolta dalla pandemia.
